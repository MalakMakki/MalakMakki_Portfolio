#Importing Librairies

```python
import numpy as np
import pandas as pd
```

#Importing dataset

```python
df=pd.read_csv("Diabetes.csv")
df
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

#Data preprocessing

##Missing values

```python
df.isnull().sum()
```

```python
df
```

**Note**: No missing values in this dataset

##Categorical values

```python
x=pd.get_dummies(x).drop(['Gender_Male','Blood Pressure_High','Family History of Diabetes_No','Smoking_No','Diet_Poor','Exercise_No'], axis=1)
x.head(5)
```

```python
x.shape
```

```python
# Normalization of data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x
```

```python
y.head(5)
```

```python
Diagnosis= {'No':0, 'Yes':1}
y = y.apply(lambda x: Diagnosis[x])
y.head()
```

```python
y
```

#Spliting Dataset

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train
```

```python
y_test
```

#Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model_Logistic = LogisticRegression()
model_Logistic.fit(x_train, y_train)
```

##Prediction

```python
y_pred_Logistic = model_Logistic.predict(x_test)
print(y_pred_Logistic)
```

```python
print(y_test)
```

##Evaluation

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(accuracy_score(y_test, y_pred_Logistic)*100)
print(confusion_matrix(y_test, y_pred_Logistic))
```

#Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(x_train, y_train)
```

##Prediction

```python
y_pred_decision_tree = classifier.predict(x_test)
print(y_pred_decision_tree)
```

```python
y_test
```

##Evaluation

```python
print(accuracy_score(y_test, y_pred_decision_tree)*100)
print(confusion_matrix(y_test, y_pred_decision_tree))
```

#Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
classifier2 = RandomForestClassifier(n_estimators=100, random_state=1)
classifier2.fit(x_train, y_train)
```

##Prediction

```python
y_pred_random_forest = classifier2.predict(x_test)
print(y_pred_random_forest)
```

```python
print(y_test.values)
```

##Evaluation

```python
print(accuracy_score(y_test, y_pred_random_forest)*100)
print(confusion_matrix(y_test, y_pred_random_forest))
```

#XGBoost

```python
from xgboost import XGBClassifier
classifier3 = XGBClassifier()
classifier3.fit(x_train, y_train)
```

##Prediction

```python
y_pred_xgboost = classifier3.predict(x_test)
print(y_pred_xgboost)
```

```python
print(y_test)
```

##Evaluation

```python
print(accuracy_score(y_test, y_pred_xgboost)*100)
print(confusion_matrix(y_test, y_pred_xgboost))
```
