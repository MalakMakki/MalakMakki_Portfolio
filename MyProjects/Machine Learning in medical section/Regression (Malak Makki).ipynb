I will build a model to predict the **cost of treatment** for individuals based on their age, sex, bmi and other information.

Dataset description:




*   age: age of primary beneficiary

*   sex: insurance contractor gender, female, male

*   bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height, objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

*   children: Number of children covered by health insurance / Number of dependents

*   smoker: Smoking

*   region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

*   **charges: Individual medical costs billed by health insurance** (target)




#Importing Libraries

```python
import numpy as np
import pandas as pd
```

#Importing Dataset

```python
df=pd.read_csv("medical_cost.csv")
df
```

#Data Understanding

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

#Data Preprocessing

##Missing value

```python
df.isnull().sum()
```

We have only 3 missing values in bmi colomn compared to 1335 present values, so we can drop rows with these missing values because they don't affect the big number of data.
**Note**: if we would not drop rows, we can visualise the distribution of bmi to know by wich value (mean, mode, median) we should replace the missing one







```python
df = df.dropna()
df.isnull().sum()
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.shape
```

##Categorical features

```python
df=pd.get_dummies(df).drop(['sex_female','smoker_no','region_southwest'], axis=1)
df.head(5)
```

```python
df.shape
```

**Note**: Values in my dataset are not very far from each other, so i would not do the normalization of values

#Data Spliting

```python
x = df[['age','bmi','children','sex_male','smoker_yes','region_northeast','region_northwest','region_southeast']]
y = df.iloc[:, 3]
```

```python
x.head()
```

```python
x.shape
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

#Linear Regression model

```python
from sklearn.linear_model import LinearRegression
r1 = LinearRegression()
r1.fit(x_train,y_train)
```

##Prediction

```python
y_pred = r1.predict(x_test)
```

```python
y_test
```

```python
y_pred.reshape(-1,1)
```

##Evaluation

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```

#Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
r2= DecisionTreeRegressor()
r2.fit(x_train,y_train)
```

##Prediction

```python
y_pred_decision_tree = r2.predict(x_test)
```

```python
y_test
```

```python
y_pred_decision_tree.reshape(-1,1)
```

##Evaluation

```python
print(r2_score(y_test, y_pred_decision_tree))
```

#Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
r3 = RandomForestRegressor(n_estimators=150)
r3.fit(x_train, y_train)
```

##Precdiction

```python
y_pred_random_forest = r3.predict(x_test)
```

```python
y_test
```

```python
y_pred_random_forest.reshape(-1,1)
```

##Evaluation

```python
print(r2_score(y_test, y_pred_random_forest))
```

#XGBoost

```python
from xgboost import XGBRegressor
r4 = XGBRegressor()
r4.fit(x_train, y_train)
```

##Prediction

```python
y_pred_xgboost = r4.predict(x_test)
```

```python
y_test
```

```python
y_pred_xgboost.reshape(-1,1)
```

##Evaluation

```python
print(r2_score(y_test, y_pred_xgboost))
```

#Conclusion

After doing thes 4 models for medical_cost dataset we can see that:


*   Linear Regression Model has a R2 score: 0.7578912886388554

*   Decision Tree Model has a R2 score: 0.6975601083504588

*   Random Forest Model has a R2 score: 0.8417671054977912

*   XGBoost Model has a R2 score: 0.8221803767859728

So we can conclude that the best model to be adapted for a great prediction of the medical cost is the Model that have the nearest R2 score to 1. So the best model is **Random Forest**
